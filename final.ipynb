```python
#importing librarires
from IPython.display import display
import pandas as pd
from datetime import datetime
from  datetime import timedelta,time
file=pd.read_excel("Assignment_Timecard.xlsx")
```

```python
#sort semi is new list containing id,date,hours,name.It does not specify the time,just every shift and the day it took place
sort_semi=[]
for row in file.iterrows():
    if(pd.isnull(row[1]["Time"])==False):
      sort_semi.append({
              'id':row[1]["Position ID"],
              'date':pd.to_datetime(row[1]["Time"]).date(),
              'hours':datetime.strptime(str(row[1]["Timecard Hours (as Time)"]),"%H:%M").time(),
              'name':row[1]['Employee Name']
          })
```

```python
#initial configuration
temp=sort_semi[0]
actual_temp={}
sort_final=[]
#logic:if the id and dates in adjacent rows are the same,add the hours,if they are different,append the 
#id,date,total hours in that day and the name 
#in the sort_final list
#final list contains only a single day and the hours worked during that day

for item in sort_semi:
    if (temp['id']!=item['id'] or temp['date']!=item['date']):
       sort_final.append(actual_temp)
       temp={
        'id':item['id'],
        'date':item['date'],
        'hours':item['hours'],
        'name':item['name']
        }
    else:
      s=temp['hours']
      c=item['hours']
      j=timedelta(hours=s.hour,minutes=s.minute)+timedelta(hours=c.hour,minutes=c.minute)
      actual_temp={
        'id':item['id'],
        'date':item['date'],
        'hours':j,
        'name':item['name']
      }
```

**People who have worked consistently for 7 days**

```python
#initial configuration
temp=sort_final[0]
consistent_7=[]
counter=0

#Assumption:The assumption made is that the employee has to work 7 days in a row
#logic:
for item in sort_final:
    if(item['id']==temp['id']):
        if(item['date']-temp['date']<=timedelta(days=1)):
            #if the difference in days in <=1 then counter increases
            counter=counter+1
            temp=item
            if(counter==7):
                #if counter reaches 7 i.e a week, the name and id is appended
                child={
                    'id':item['id'],
                    'name':item['name']
                }
                if child in consistent_7:
                    pass
                else:
                    consistent_7.append(child)
        else:
            #if the streak is broken , counter is reset
            counter=1
            temp=item
    else:
        #if the person switches,counter is reset 
        temp=item
        counter=1
```

**Checking people who have worked for more than 1 hour and less than 10 hour between shifts**

```python
#initial configuration
flag=0
temp=sort_final[0]
upper=timedelta(days=0,hours=10)
lower=timedelta(days=0,hours=1)
b_final=[]
#Assumption:Assuming the problem asks to find people who have worked between 1 to 10 hours in total between all shifts for the day.

#logic:Flag variable 0 symbolises that person is satisfying the criteria,1 states otherwise.
for row in sort_final:
  #if flag is 0 and adjacent rows have different id's,i.e previous person has worked all days between 1 to 10 hours
  if(flag==0 and temp['id']!=row['id']):
        child={
        'id':temp['id'],
        'name':temp['name']
        }
        temp=row
        #append the person to final list
        if child in b_final:
          pass
        else:
          b_final.append(child)
  elif(flag==0):
    #this is if flag is 0 but person remains the same,to make sure that all the days satisfy the criteria
    if(row['hours']<upper and row['hours']>lower):
      temp=row
    else:
      flag=1
      temp=row
  elif(flag==1 and temp['id']!=row['id']):
    #if person switches and previous person didnt satisfy the condition, then flag is reset 
    flag=0
    if(row['hours']<upper and row['hours']>lower):
      temp=row
    else:
      flag=1
      temp=row
```

**People who have worked for more than 14 hours in one shift**

```python
#initial configuration
c_final=[]

#Logic:Go through every single shift and check which is greater than 14 hours
for row in sort_semi:
  if(row['hours']>time(14,0,0)):
    child={
        'id':row['id'],
        'name':row['name']
        }
    if child in c_final:
      pass
    else:
      c_final.append(child)
```
